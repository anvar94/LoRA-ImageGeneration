```python
from diffusers import StableDiffusionPipeline
from huggingface_hub.repocard import RepoCard
import torch
import matplotlib.pyplot as plt

# Define your LoRA model (update this to match your use case)
use_hub = True  # Set to False if using a local path
lora_model_id = "Anvar94/output3"  # Replace with Hugging Face repo ID or local path
base_model_id = "stabilityai/stable-diffusion-v1-5"  # Base model

if use_hub:
    try:
        # Load model card to fetch base model dynamically (if available)
        card = RepoCard.load(lora_model_id)
        if "base_model" in card.data.to_dict():
            base_model_id = card.data.to_dict()["base_model"]
    except Exception as e:
        print(f"Error loading LoRA model card: {e}")
        print("Please verify the repository exists or use a local path instead.")
        exit(1)

print(f"Base model ID!!!!!!: {card}")
# Load Stable Diffusion pipeline
pipe = StableDiffusionPipeline.from_pretrained(base_model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# Load LoRA weights (try using hub or local path)
try:
    pipe.load_lora_weights(lora_model_id)
except Exception as e:
    print(f"Error loading LoRA weights: {e}")
    print("Ensure the LoRA weights are uploaded to the Hub or provide a valid local path.")
    exit(1)

print("Pipeline and LoRA weights loaded successfully!")
```

```python
# Prompt 1
prompt = "A picture of fogi bear sitting on a beach at sunset"
num_inference_steps = 25

# Generate Image
image = pipe(prompt, num_inference_steps=num_inference_steps).images[0]

# Display the Image
plt.imshow(image)
plt.axis("off")
plt.title(f"Prompt: {prompt}", fontsize=12)
plt.show()
```
